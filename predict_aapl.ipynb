```python
aapl = yf.Ticker("AAPL")
aapl.info["forwardPE"]
```

```python
aapl = aapl.history(period="max")
```

```python
aapl["Tomorrow"] = aapl["Close"].shift(-1)
```

```python
aapl
```

```python
aapl["Target"] = (aapl["Tomorrow"]>aapl["Close"]).astype(int)
```

```python
aapl
```

```python
aapl = aapl.loc["2000-01-01":].copy()
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = aapl.iloc[:-100]
test= aapl.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score

predictions = model.predict(test[predictors])
```

```python
import pandas as pd 

predictions = pd.Series(predictions, index=test.index)
```

```python
precision_score(test["Target"], predictions)
```

```python
def predict(train, test, predictors, model): 
        model.fit(train[predictors], train["Target"])
        predictions = model.predict(test[predictors])
        predictions = pd.Series(predictions, index=test.index, name = "Predictions")
        combined = pd.concat([test["Target"], predictions], axis=1)
        return combined
```

```python
def backtest(data, model,predictors,start=1000,step=250):
    all_predictions = []
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(aapl, model, predictors)
```

```python
precision_score(predictions["Target"],predictions["Predictions"])
```

```python
horizons = [2, 5, 60, 250, 1000]

new_predictors = []

for horizon in horizons:
    rolling_averages = aapl.rolling(horizon).mean()
    
    ratio_column = f"Close Ratio_{horizon}"
    aapl[ratio_column] = aapl["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    aapl[trend_column] = aapl.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors += [ratio_column, trend_column]
```

```python
aapl = aapl.dropna()
```

```python
import xgboost as xgb
from sklearn.model_selection import StratifiedKFold
import numpy as np
```

```python
n_splits = 5
kf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
```

```python
X = aapl.drop(columns=["Target"])  # Features
y = aapl["Target"]  # Target
```

```python
n_splits = 5
kf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
models = []

for fold, (train_idx, val_idx) in enumerate(kf.split(X, y)):
    print(f"\n<== Training fold {fold + 1}/{n_splits} ==>")
        
    # Split into train and validation sets for the fold
    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
    
    # Initialize the XGBoost model
    model = xgb.XGBClassifier(
        n_estimators=5000,
        learning_rate=0.03,
        max_depth=10,
        min_child_weight=8,
        subsample=0.85,
        colsample_bytree=0.5,
        reg_alpha=0.8,
        reg_lambda=0.4,
        random_state=2005,
        eval_metric="auc",
        tree_method="hist",  # fast and good for tabular
        use_label_encoder=False
    )
    
    # Fit the model on the training data
    model.fit(X_train, y_train, eval_set=[(X_val, y_val)], verbose=100)
    
    # Run backtest on the full dataset (not just on the fold)
    predictions = backtest(aapl, model, new_predictors)
    
    # Calculate precision score on the predictions from backtest
    precision = precision_score(predictions["Target"], predictions["Predictions"])
    print(f"Precision Score for Fold {fold + 1}: {precision}")
    
    # Store the trained model
    models.append(model)

print("\nXGBoost model training complete.")
```

```python
def predict(train, test, predictors, model): 
        model.fit(train[predictors], train["Target"])
        predictions = model.predict_proba(test[predictors])[:,1]
        predictions[predictions >= .6] = 1
        predictions[predictions < .6] = 0
        predictions = pd.Series(predictions, index=test.index, name = "Predictions")
        combined = pd.concat([test["Target"], predictions], axis=1)
        return combined
```

```python
predictions = backtest(aapl, model, new_predictors)
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
